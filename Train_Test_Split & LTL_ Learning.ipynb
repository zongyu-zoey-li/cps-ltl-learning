```python
import os
import numpy as np
import shutil
import sklearn
from sklearn.model_selection import train_test_split
from Scarlet.ltllearner import LTLlearner
```

```python
root= '../Scarlet-main/All_benchmarks/RQ1-benchmarks/generated_benchmarks/TracesFiles'
save_root = './'
```

```python
for i in range(1, 9):
    file_list = []
    
    with open(os.path.join(os.path.dirname(root), 'formulas.txt'), 'r') as f:
        Lines = f.readlines()
    for j, line in enumerate(Lines):
        if j+1 == i:
            LTL_formula = line
            break
    
    for file in os.listdir(root):
        if file.startswith('f:0{}'.format(i)):
            file_list.append(file)
    train, test = train_test_split(file_list, test_size=0.2)
    save_root_formula = os.path.join(save_root, 'formula{}'.format(i))
    if not os.path.exists(save_root_formula):
        os.makedirs(save_root_formula)
    
    with open(os.path.join(save_root_formula, 'formula{}.txt'.format(i)), 'w') as f:
        f.write(LTL_formula)
    f.close()
    
    dest_train = os.path.join(save_root_formula, 'train')
    dest_test = os.path.join(save_root_formula, 'test')
    if not os.path.exists(dest_train):
        os.makedirs(dest_train)
    if not os.path.exists(dest_test):
        os.makedirs(dest_test)
    
    train_file = []
    test_file = []
    for file in os.listdir(root):
        if file in train:
            shutil.copy(os.path.join(root, file), dest_train)
        elif file in test:
            shutil.copy(os.path.join(root, file), dest_test)
```

```python
for i in range(1, 9):
    formula_num = i
    train_root = './formula{}/train'.format(formula_num)
    for file in os.listdir(train_root):
        file_path = os.path.join(train_root, file)
        learner = LTLlearner(input_file = file_path, timeout=900, csvname='formula{}.xlsx'.format(formula_num))
        learner.learn()
```
